```python
import pandas as pd
```

# DataFrame行操作

## 添加行，会返回新的DataFrame

### 加载scientists数据集

```python
scientists=pd.read_csv('./data(1)/scientists.csv')
scientists
```

### 实例：在scientists数据末尾添加一行新数据

```python
# 准备新行数据
new_series=pd.Series(['LuoGeng Hua', '1910-11-12', '1985-06-12', 75, 'Mathematician'],
index=['Name', 'Born', 'Died', 'Age', 'Occupation'])
scientists.append(new_series,ignore_index=True)
```

## 修改行

### 修改行时，直接对原始DataFrame操作

```python
# 修改行标签为4的行的所有数据
scientists.loc[4]=['Rachel Carson','1907/05/27','1964/04/14','56','Biologist']
scientists
```

```python
scientists.iloc[4]=['Rachel Carson','1907/05/27','1964/04/14','56','Biologist']
scientists
```

### 修改行标签为4行的Born和Age列的数据

```python
scientists.loc[4,['Born','Age']] = ['1906-5-27',58]
scientists
```

### 示例：修改行标签为6的行Born列的数据为1912-07-23

```python
scientists.loc[6,['Born']] = '1912-07-23'
scientists
```

## 删除行

### 删除行时会返回新的DataFrame

```python
#删除行标签为1和3的行
scientists.drop([1,3])
```

# DataFrame列操作

## 新增列、修改列

```python
#注意：添加列/修改列时，是直接对原始的DataFrame进行修改
```

```python
scientists['Country']=['England', 'England', 'England', 'French','America', 'England', 'England', 'Germany']
scientists
```

```python
#实例：给scientists数据增加一个Country列
scientists.loc[:,'Country']=['England', 'England', 'England', 'French','America', 'England', 'England', 'Germany']
scientists
```

```python
#示例：修改scientists数据中Country列的数据
scientists['Country']=['england', 'england', 'england', 'french','america', 'england', 'england', 'germany']
scientists
```

## 删除列

###  注意：删除列时，会返回一个新的DataFrame

```python
#示例：删除scientists数据中Country列的数据
scientists.drop(['Country'],axis=1)
```

# DataFrame数据导出和导入

## Pickle 文件

### 导出到Pickle文件

```python
scientists.to_pickle('./scientists_df.pickle')
```

### 读取pickle文件

```python
scientists_df=pd.read_pickle('./scientists_df.pickle')
scientists_df
```

## CSV文件

```python
#tsv文件，设置分隔符必须是\t
scientists.to_csv('./scientists_df.csv',sep='\t')
```

```python
# 不在CSV文件中存入行标签
scientists.to_csv('./scientists_df_noindex.csv',index=False)
```

## Excel文件

### 导出成Excel文件

```python
scientists.to_excel('./scientists_df.xls',sheet_name='scientists',index=False)
```

### 读取Excel文件：

```python
pd.read_excel('./scientists_df.xls')
```
