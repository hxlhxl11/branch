# 数据组合简介

# concat 拼接数据

##  行拼接：按照列标签索引对齐

```python
# DataFrame和DataFrame 进行拼接：
# 注意：行拼接时，无法对齐的列，默认在拼接后的数据中会填充为 NaN，因为默认拼接是 outter 拼接
```

### 加载数据集

```python
df1=pd.read_csv('./data(1)/concat_1.csv')
df1
```

```python
df2=pd.read_csv('./data(1)/concat_2.csv')
df2
```

```python
df3=pd.read_csv('./data(1)/concat_3.csv')
df3
```

### 将 df1、df2和 df3 按照列标签对齐，进行行拼接

```python
row_concat =pd.concat([df1,df2,df3])
row_concat
```

```python
# 按行位置获取数据（按索引来）
row_concat.iloc[5]
```

```python
# 按行标签来获取数据
row_concat.loc[3]
```

### concat拼接时忽略原有的行标签、

```python
pd.concat([df1,df2,df3],ignore_index=True)
```

## DataFrame 和 Series 进行拼接：

### 创建一个series数据

```python
new_series=pd.Series(['n1','n2','n3','n4'])
print(new_series)
```

### 将 df1 和 new_series 按照列标签对齐，进行行拼接

```python
pd.concat([df1,new_series])
```

## 列拼接：按照行标签索引对齐

### DataFrame和DataFrame 进行拼接：

```python
# 将 df1、df2和 df3 按照行标签对齐，进行列拼接
pd.concat([df1,df2,df3],axis=1)
```

### DataFrame 和 Series 进行拼接：

```python
# df1 和 new_series 按照行标签对齐，进行列拼接
pd.concat([df1,new_series],axis=1)
```

## Join 的参数设置

```python
#concat 方法的 join参数：
#默认为 outter：无法对齐的行/列，默认在拼接后的数据中会填充为 NaN，因为默认拼接是 outter 拼接
#设置为 inner：只有能够对齐的行/列，才会出现在拼接的结果中
```

### 修改df1和df3的columns列标签

```python
df1.columns=['A','B','C','D']
df1
```

```python
df3.columns=['A','C','F','H']
df3
```

### 将 df1和df3按列标签对齐，进行行拼接，默认outter

```python
pd.concat([df1,df3])
```

### 将 df1 和 df3 按列标签对齐，进行行拼接，设置 join='inner'

```python
pd.concat([df1,df3],join='inner')
```

# merge 关联数据

## 方法简介
merge 方法类似于 sql 中的 join 语句，用于两个数据集之间按照行标签索引|列标签索引链接，默认是inner，可以设置为：left、right、outer
